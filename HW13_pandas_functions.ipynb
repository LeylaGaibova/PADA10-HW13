###### Задание 1**

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
ratings = pd.read_csv('D:\\Python\\Lesson12_Функции и работа с данными_290620\\zip\\ratings.csv')
```

```python
ratings.head(2)
```

```python
def classification(param):
    if param <= 2:
        rating_class = 'низкий рейтинг'
    elif param <= 4:
        rating_class = 'средний рейтинг'
    elif param == 4.5 or param == 5:
        rating_class = 'высокий рейтинг'
    return rating_class
```

```python
ratings['class'] = ratings.rating.apply(classification)
ratings
```

###### Задание 2

Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

Результат классификации запишите в отдельный столбец region

```python
keywords = pd.read_csv('D:\\Python\\Lesson12_Функции и работа с данными_290620\\keywords.csv')
```

```python
keywords.tail()
```

```python
def geo_classificator(string):
    str = ''
    for key, value in geo_data.items():
        for word in value:
            if word in string:
                str = key
    if str == '':
        str = 'undefined'
    return str
```

```python
keywords['region'] = keywords['keyword'].apply(geo_classificator)
keywords.head(3)
```

```python
not_undefined = keywords[keywords.region != 'undefined']
not_undefined.head(20)
```

Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

В переменную years запишите список из всех годов с 1950 по 2010.

Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

для каждой строки пройдите по всем годам списка years
если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

```python
ratings1 = pd.read_csv('D:\\Python\\Lesson12_Функции и работа с данными_290620\\zip\\ratings.csv')
```

```python
movies1 = pd.read_csv('D:\\Python\\Lesson12_Функции и работа с данными_290620\\zip\\movies.csv')
```

```python
joined = ratings1.merge(movies1, on='movieId', how='left').filter(items = ['movieId', 'title', 'rating'])
```

```python
def production_year(param):
    years  = [i for i in range(1950,2011)]
    result = 0
    for year2 in years:
        if str(year2) in param:
            result = year2
    if result == 0:
        result = 1900
    return result
```

```python
joined['year'] = joined.title.apply(production_year)
```

```python
pd.set_option('max_rows', 62)
```

```python
joined = joined.filter(items = ['year', 'rating'])
joined.groupby(['year']).mean().sort_values('rating', ascending=False)
```
